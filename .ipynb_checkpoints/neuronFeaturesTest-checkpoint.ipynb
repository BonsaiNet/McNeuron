```python
j = time.time()
import pandas as pd
neurons = pd.read_pickle("/Users/anirudhnatarajan/Desktop/neuronData/neurons.csv")
f = time.time()
print(int(datetime.datetime.fromtimestamp(f).strftime('%S')) - int(datetime.datetime.fromtimestamp(j).strftime('%S')))
```

```python
size = 100
n = [0] * size
```

```python
for i in range (0,size) :
    r = randint(0,79992)
    neuron = McNeuron.Neuron(neurons['swc'][r])
    neuron.l_measure_features()
    n[i] = neuron.getNeuronVector()
    f = time.time()
    if i%10==0 :
        print(i)
    t = int(datetime.datetime.fromtimestamp(f).strftime('%S')) - int(datetime.datetime.fromtimestamp(j).strftime('%S'))
    if t>10 | t<-50:
        print(t,r)
print('done')
```
