```python
import numpy as np
import McNeuron
import matplotlib.pyplot as plt
from copy import deepcopy
from numpy.linalg import inv
from collections import Counter
import pickle
%matplotlib inline
```

## Plotting neuron

```python
neuron = McNeuron.Neuron("neuron.txt")
McNeuron.visualize.plot_2D(neuron)
```

## Plotting a part o neuron

```python
swc_matrix = neuron.get_swc()
S = McNeuron.subsample.Subsample()
S.set_swc(swc_matrix)
dendrite = S.select_part_of_neuron(part='dendrite')
McNeuron.visualize.plot_2D(dendrite)
```

## Subsampling

```python
S = McNeuron.subsample.Subsample()
S.set_swc(swc_matrix)
S.fit()
subsampled_neuron = S.subsample(subsample_type='straigthen', length=15)

McNeuron.visualize.plot_2D(subsampled_neuron)
```
